```python
#checking count of true and false in the Outcome column
Outcome_true_count = len(data.loc[data['Outcome'] == True])
Outcome_false_count =  len(data.loc[data['Outcome'] == False])
```

```python
#splitting the x and y values
from sklearn.model_selection import train_test_split
features_columns =['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
predict_class = ['Outcome']
```

```python
x = data[features_columns].values
y=data[predict_class].values
#Splitting the train and test value
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

```python
#Replacing the place with value zero with the mean
from sklearn.impute import SimpleImputer
fill = SimpleImputer(missing_values=0,strategy ="mean" )
x_train = fill.fit_transform(x_train)
x_test = fill.fit_transform(x_test)
```

```python
#Applying RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
r =  RandomForestClassifier()
r.fit(x_train,y_train)
```

```python
#predicting the value
yp = r.predict(x_test)
```

```python
#accuracy score
r.score(x_test,y_test)
```
